```python
#Implementation of Stack using Linked List
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
class Stack:
  def __init__(self):
    self.top = None
  def isempty(self):
    return self.top==None
  def push(self, element):
    newnode = Node(element)
    newnode.next = self.top
    self.top = newnode
  def pop(self):
    if self.isempty():
      print("Stack underflow")
    else:
      t = self.top.data
      self.top = self.top.next
      return t
  def top_(self):
    return self.top.data
  def display(self):
    if self.isempty():
      print("Stack underflow")
    else:
      temp = self.top
      while temp:
        print(temp.data, end=" ")
        temp = temp.next
      print()
```

```python
s=Stack()
s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.display()
s.pop()
s.display()
s.top_()
```
